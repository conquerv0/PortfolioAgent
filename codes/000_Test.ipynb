```python
import numpy as np
import pandas as pd
from google.colab import drive

def generate_and_save_random_data(filepath):
  # Generate random data
  data = np.random.rand(1000, 3)

  # Create a Pandas DataFrame
  df = pd.DataFrame(data, columns=['Column1', 'Column2', 'Column3'])

  # Save the DataFrame to a CSV file in Google Drive
  df.to_csv(filepath, index=False)

  print(df)


# Set the file path
filepath = '/content/drive/My Drive/MS Finance Thesis Shared Drive/data/test_data.csv'

# Example usage (replace with your desired file path)
drive.mount('/content/drive')
generate_and_save_random_data(filepath)
```
